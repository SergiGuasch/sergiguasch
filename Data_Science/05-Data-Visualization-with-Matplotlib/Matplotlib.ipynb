```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'y')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #The dimensions (left, bottom, width, height). The values must be betwwen 0 and 1
axes.plot(x,y,'r')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'y')
axes2.plot(y,x,'g')
axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout() #to avoid overlapping plots
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
 axes #running this, we can see is an array so it means that we can iterate through
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_xlabel('X label')
axes[0].set_ylabel('Y label')
axes[0].set_title('Title')

axes[1].plot(y,x)
axes[1].set_xlabel('X label')
axes[1].set_ylabel('Y label')
axes[1].set_title('Title')
```

```python
# Figure size and DPI

fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize = (8,2))

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
# add legend

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='X Squared')
ax.plot(x,x**3, label ='X Cubed')

ax.legend(loc = 0) # loc is to specify relocation of legend. 0 is for best location. 
                   # Also can specify with tuples such as loc=(0.1,0.1)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00') # use RGB Hex Code for colors
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', linewidth=10) # or lw
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', lw=5, alpha=0.5) #alpha is for transpharency
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red', lw=3, linestyle='--') # or ls             -    --   -.   : steps
```

```python
#markers

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', lw=0.5, linestyle='-', marker = '+', markersize = 20) # o * + 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', lw=1, linestyle='-', marker = 'o', markersize = 20, 
        markerfacecolor = 'yellow', markeredgewidth=3, markeredgecolor='red')  
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', lw=2, ls='--')

ax.set_xlim([0,1]) #to show the plot between 0 and 1 on the x axis, needs to pass a list with lower bound and upper bound
ax.set_ylim([0,2]) #to show the plot between 0 and 2 on the y axis, needs to pass a list with lower bound and upper bound
```
