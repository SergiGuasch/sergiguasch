```python
df1.plot.line(y='B', figsize=(12,3), lw=0.5) #x=df1.index not given
```

```python
df1.plot.scatter(x='A', y='B', color='C', cmap='coolwarm') #color='C' or c='C' is a third dimensional argument
```

```python
df1.plot.scatter(x='A', y='B', size='C', cmap='coolwarm') # size='C' or s='C' is a third dimensional argument and is better
                                                          # but is too small, better multiply a factor, like *100
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100, c='C', cmap='coolwarm') # also we can combine with a third and fourth argument
```
